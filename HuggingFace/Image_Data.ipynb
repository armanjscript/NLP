```python
import cv2
```

```python
img = cv2.imread('puppy.jpg')
```

```python
plt.imshow(img) #Open CV BGR  matplotlib RGB
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
#Resizing photo
img_resize = cv2.resize(img, (500, 1000))
plt.imshow(img_resize)
```

```python
w_ratio = 0.5
h_ratio = 0.5
new_image = cv2.resize(img, (0, 0), fx=w_ratio, fy=h_ratio)
```

```python
plt.imshow(new_image)
```
